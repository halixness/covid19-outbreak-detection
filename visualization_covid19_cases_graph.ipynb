```
import pandas as pd
import io
import requests
from copy import deepcopy

# Read updated dataset from source
url = "https://raw.githubusercontent.com/RamiKrispin/coronavirus-csv/master/coronavirus_dataset.csv"
s = requests.get(url).content
dataset = pd.read_csv(io.StringIO(s.decode('utf-8')))

# Group by country, get timeseries by day with cases
list = dataset.groupby('Country.Region').apply(lambda x: [list(x['date']), list(x['cases'])]).apply(list)
```

```
import numpy as np
from datetime import datetime

# Get cities
db_dict = list.to_dict()
dim = len(db_dict)

# Start Adjances Matrix
graphs = dict()
dates = []
```

## Sorting the dates for which cases have been found

```
# Getting dates
for key in db_dict:
  value = db_dict[key]
  k_dates = value[0]

  # Save all unique dates
  for date in k_dates:
    if date not in dates:
      dates.append(date)
    
# Chronological sort
dates.sort(key=lambda date: datetime.strptime(date, "%Y-%m-%d"))
```

## Populating the adjances matrices day by day

```
import matplotlib.pyplot as plt
import networkx as nx

def show_graph_with_labels(adjacency_matrix, mylabels):
  
    rows, cols = np.where(adjacency_matrix == 1)
    edges = zip(rows.tolist(), cols.tolist())
    
    gr = nx.Graph()

    gr.add_edges_from(edges)

    nx.draw(gr, node_size=500, labels=mylabels, with_labels=False)
    
    plt.show()
```

```
from IPython.display import clear_output
from time import sleep

debug = True
cap = False
dense_mode = False
```

## Classical linking approach


```
zero_cities = []
graph_series = dict()
infected_nodes = []
old_infected = []
new_infected = []

### Creating adjacency matrix for each day
j = 0

for day in dates:  

  links = 0

  # Empty matrix
  if j is 0:
    day_graph = np.zeros((len(db_dict), len(db_dict)))
  else:
    day_graph = graph_series[dates[j-1]]

  ### Getting new infected connections
  x = 0

  if debug:
    print("\n****************** j = " + str(j) + " ********************")

    print("Old infected nodes: ")
    print(old_infected)
    print("New infected nodes: ")
    print(new_infected)
    print((new_infected == old_infected))
  

  # if we actually have new cases
  if not (new_infected == old_infected):
    
    # Previous new cities -> influence new ones
    if not dense_mode:
      for old_inf in old_infected:
        for new_inf in new_infected:

          # Bidirectional
          day_graph[new_inf][old_inf] = 1
          day_graph[old_inf][new_inf] = 1
          
          if debug:
            print(str(new_inf) + " --- " + str(old_inf))

    # Every infected city can influence new nodes
    else:
      for old_inf in infected_nodes:
        for new_inf in new_infected:

          # Bidirectional
          day_graph[new_inf][old_inf] = 1
          day_graph[old_inf][new_inf] = 1
          
          if debug:
            print(str(new_inf) + " --- " + str(old_inf))


  print("\n")

  show_graph_with_labels(np.asarray(day_graph), [])

  # First 15 days cap
  if j == 15 and debug and cap:
    raise KeyboardInterrupt

  graph_series[day] = day_graph

  ### Get zero cities
  x = 0
  for record in db_dict:
    r_dates2 = db_dict[record][0]

    # There are registered cases for that day
    if day not in r_dates2:
      zero_cities.append(x)
    
    # There are 0 cases for that day
    else:
      index = r_dates2.index(day)
      if db_dict[record][1][index] == 0:
         zero_cities.append(x)

    x += 1

  ### Getting infected cities
  x = 0

  if len(new_infected) is not 0:
    old_infected = new_infected

  # Reset new infected
  new_infected = []

  for record in db_dict:
    r_dates = db_dict[record][0]
    index = r_dates.index(day)
    cases = db_dict[record][1][index]

    # There are registered cases for that day
    if day in r_dates and cases is not 0:
      
      # If it's a new city
      if x not in infected_nodes:
        new_infected.append(x)
        infected_nodes.append(x)

    x += 1

  j += 1 
```
